```python
import pandas as pd
import numpy as np
import unicodedata

# Read the excel file with plants  information
file_path = r'C:\Users\NW\Documents\Computer Science\Yachay Tech Major\Sexto semestre\Data Bases\medicinal_plants_database\Medicinal-plants-proyect\Data_cleaning\Proyecto plantas medicinales_test.xlsx'
excel_data = pd.ExcelFile(file_path)

locations = {'Imbabura': ["Antonio Ante", "Urcuqui", "Pimapiro", "Otavalo", "Cotacachi", "Ibarra"]}

def create_dataframe_info_antonio_ante(excel_data):
    antonio_ante_excel_dataframe = pd.read_excel(excel_data, sheet_name='Antonio Ante')
    antonio_ante_columns_general_info_db = ['nombre_comun']

    antonio_ante_nombre_comun_data = antonio_ante_excel_dataframe.loc[1:, 'Nombre Común ']
    antonio_ante_nombre_cientifico_data = antonio_ante_excel_dataframe.loc[1:, 'Nombre Científico']
    antonio_ante_uso_ancestral_reportado = antonio_ante_excel_dataframe.loc[1:, 'Uso']
    antonio_ante_uso_ancestral_fuentes_vivas = antonio_ante_excel_dataframe.loc[1:, 'Unnamed: 5']
    antonio_ante_uso_comprobado = antonio_ante_excel_dataframe.loc[1:, 'Unnamed: 6']
    antonio_ante_origen = antonio_ante_excel_dataframe.loc[1:, 'Toponimo ']
    antonio_ante_distribucion = antonio_ante_excel_dataframe.loc[1:, 'Información de procedencia (De donde lo traen, o si es cultivo propio)']
    antonio_ante_distribucion_comercial = antonio_ante_excel_dataframe.loc[1:, 'Lugar de distribución']
    antonio_ante_referencia_1 = antonio_ante_excel_dataframe.loc[1:, 'Investigaciones previas']
    antonio_ante_referencia_2 = antonio_ante_excel_dataframe.loc[1:, 'Unnamed: 11']
    antonio_ante_referencia_3 = antonio_ante_excel_dataframe.loc[1:, 'Unnamed: 12']
    
    antonio_ante_dataframe = pd.DataFrame(data=list(antonio_ante_nombre_comun_data), columns=antonio_ante_columns_general_info_db)

    antonio_ante_dataframe['nombre_cientifico'] = list(antonio_ante_nombre_cientifico_data)
    antonio_ante_dataframe['ciudad'] = ["Antonio Ante"] * len(antonio_ante_dataframe)
    antonio_ante_dataframe['provincia'] = ["Imbabura"] * len(antonio_ante_dataframe)
    antonio_ante_dataframe['origen'] = list(antonio_ante_origen)
    antonio_ante_dataframe['distribucion'] = list(antonio_ante_distribucion)
    antonio_ante_dataframe['distribucion_comercial'] = list(antonio_ante_distribucion_comercial)
    antonio_ante_dataframe['uso_ancestral_reportado'] = list(antonio_ante_uso_ancestral_reportado)
    antonio_ante_dataframe['uso_ancestral_segun_fuentes_vivas'] = list(antonio_ante_uso_ancestral_fuentes_vivas)
    antonio_ante_dataframe['uso_cientificamente_comprobado'] = list(antonio_ante_uso_comprobado)
    antonio_ante_dataframe['referencia_1'] = list(antonio_ante_referencia_1)
    antonio_ante_dataframe['referencia_2'] = list(antonio_ante_referencia_2)
    antonio_ante_dataframe['referencia_3'] = list(antonio_ante_referencia_3)
    
    return antonio_ante_dataframe

antonio_ante_excel_dataframe = pd.read_excel(excel_data, sheet_name='Antonio Ante')

# print(antonio_ante_excel_dataframe.columns)
# print(antonio_ante_excel_dataframe.loc[0:, 'Unnamed: 12'])


antonio_ante_dataframe = create_dataframe_info_antonio_ante(excel_data)

antonio_ante_dataframe.head(3)
```
